```python
import pandas as pd
import numpy as np

#this is just the code that created zeroed_corners.list. I would not recommend re-running
#it as the data has already been collected, but if you do, go into the file and make
#sure the column names are correct -- there's a bug in the csv file save

# List to store data from columns 3-11
data = []

```

```python
# Iterate over file numbers 1-62
data =[]
for i in range(1, 63):
    #filename that contains corners for a given exposure
    filename = f"/data/des80.a/data/desgw/ImageProcessingTesting/gw_workflow/669850_local/42/974612_{i}.out"
    

    try:
        # Read the file
        with open(filename, 'r') as file:
            line = str(file.readline().strip())
            #print(f'line is {line}')
            # Split the line by spaces
            values = line.split()
#             print(f'values are {values}')

            data.append(values)
#             print(f'data is {data}')
        
    except FileNotFoundError:
        print(f'hi {filename}')
        # If the file is not found, continue to the next iteration
        #should detect ccd 2 and 61 are missing
        continue
        
# print(data)
```

```python
def find_exposure(file_path, exposure_number):
    #getting ra and dec for this exposure for rezeroing
    try:
        with open(file_path, 'r') as file:
            for line in file:
                if exposure_number in line:
                    return line.strip()
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    
    return None

# Example usage
exposure_file = "/data/des80.a/data/eliseke/gw_workflow/exposures.list"
target_exposure = "974612"

found_line = find_exposure(exposure_file, target_exposure)

if found_line:
    print("Exposure found:")
    print(found_line)
else:
    print("Exposure not found.")
```

```python
# Concatenate all data frames into a single data frame
# Create a data frame from the collected data
result = pd.DataFrame(data)
# result.rename(index={0: "", 1: "y", 2: "z"})
# print(result.columns)
result.rename(columns={0: "Exposure", 1: "Band", 2: "CCD", 
                       3: "Corner1 Ra", 4: "Corner1 Dec", 
                       5: "Corner2 Ra", 6: "Corner2 Dec", 
                       7: "Corner3 Ra", 8: "Corner3 Dec",
                       9: "Corner4 Ra", 10: "Corner4 Dec"})
# Print the final result
# df.rename(columns={"A": "a", "B": "c"})
print(result.iloc[[1]])
expra = 195.555196
expdec = -23.265400
```

```python
CCDs = []
CCD_rz1 = []
CCD_rz2 = []
CCD_rz3 = []
CCD_rz4 = []
CCD_rz5 = []
CCD_rz6 = []
CCD_rz7 = []
CCD_rz8 = []

#zero the corners 
for i in range(len(result)):
    CCDs.append(float(result[2][i]))
    CCD_rz1.append((float(result[3][i])-expra)*np.cos(expdec*2*np.pi/360))
    CCD_rz3.append((float(result[5][i])-expra)*np.cos(expdec*2*np.pi/360))
    CCD_rz5.append((float(result[9][i])-expra)*np.cos(expdec*2*np.pi/360))
    CCD_rz7.append((float(result[7][i])-expra)*np.cos(expdec*2*np.pi/360))
    CCD_rz2.append(float(result[4][i])-expdec)
    CCD_rz4.append(float(result[6][i])-expdec)
    CCD_rz6.append(float(result[10][i])-expdec)
    CCD_rz8.append(float(result[8][i])-expdec)
# print(CCD_rz1)
```

```python
import csv
#save a file that shows CCD and then all corners
def lists_to_csv(lists, column_names, file_path):
    # Transpose the lists to convert rows into columns
    print(lists)
    columns = list(zip(*lists))
    print(columns)
    # Write the transposed lists to a CSV file
    with open(file_path, 'w', newline='') as file:
        writer = csv.writer(file)
        
        # Write column names as the first row
        writer.writerow(column_names)
        
        # Write the data rows
        for column in columns:
            print(column)
            writer.writerow(column)
        
lists = [CCDs, CCD_rz1, CCD_rz2, CCD_rz3, CCD_rz4, CCD_rz5, CCD_rz6, CCD_rz7, CCD_rz8]
column_names = ['CCD', "Corner 1 Ra", "Corner 1 dec", "Corner 2 Ra", "Corner 3 Dec", "Corner 4 Ra", "Corner 4 Dec"]
output_file = 'zeroed_corners.list'
```

```python

lists_to_csv(lists, column_names, output_file)
print(f"CSV file '{output_file}' has been created.")
```
